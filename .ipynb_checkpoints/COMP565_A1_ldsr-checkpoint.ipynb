## Solution

#### Given data is stored under following structure: `./data/Q1/`

```python
import pandas as pd
```

```python
M = 4268 # Number of SNPs
N = 1000 # Number of Individuals
```

```python
# Please notice the path difference, also I was getting some errors loading gzip directly so, I am using uncompressed versions of file.
# If you are using linux you can use `gzip -d filename` to uncompress files.
LD_pearson_corr = pd.read_csv('data/Q1/LD.csv', index_col=0)
```

```python
# Please notice the path difference, also I was getting some errors loading gzip directly so, I am using uncompressed versions of file.
# If you are using linux you can use `gzip -d filename` to uncompress files.
beta_marg = pd.read_csv('data/Q1/beta_marginal.csv', index_col=0)
```

```python
LD_pearson_corr
```

```python
beta_marg
```

We can solve this regression problem through multiple way, we will use the simplest and intuitive analytical way of using calculus. Motivations for this approach can be found here: https://web.ma.utexas.edu/users/m408m/Display14-7-5.shtml

```python
summary_stats = pd.DataFrame()
```

```python
summary_stats['LD_score'] = LD_pearson_corr.apply(lambda row: (row**2).sum(), axis=1)
```

```python
summary_stats['beta'] = beta_marg['V1']**2
```

```python
summary_stats
```

```python
sum_beta_LD = (summary_stats['beta'] * summary_stats['LD_score']).sum()
```

```python
sum_LD = summary_stats['LD_score'].sum()
```

```python
sum_LD_squared = (summary_stats['LD_score']**2).sum()
```

```python
heritability = M * ((N * sum_beta_LD - sum_LD) / (N * sum_LD_squared))
```

```python
print("Estimated Heritability: ", heritability)
```
