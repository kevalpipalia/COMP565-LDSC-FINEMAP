### Question 1 [2%] Implementing LD score regression

```python
import pandas as pd
```

```python
LD_pearson_corr = pd.read_csv('data/LD.csv', index_col=0)
```

```python
beta_marg = pd.read_csv('data/beta_marginal.csv', index_col=0)
```

```python
LD_pearson_corr
```

```python
summary_stats = pd.DataFrame(LD_pearson_corr.apply(lambda row: (row ** 2).sum(), axis=1))
```

```python
summary_stats.columns = ['LD_r_squared']
```

```python
summary_stats['beta_marginal'] = beta_marg['V1'] ** 2
```

```python
summary_stats
```

```python
M = 4268 # Number of SNPs
N = 1000 # Number of Individuals
```

```python
LD_r_squared_mean = summary_stats['LD_r_squared'].values.mean()
```

```python
LD_r_squared_mean
```

```python
beta_marginal_mean = summary_stats['beta_marginal'].values.mean()
```

```python
beta_marginal_mean
```

```python
summary_stats['diff_beta'] = summary_stats['beta_marginal'] - beta_marginal_mean
```

```python
summary_stats['diff_LD_r_squared'] = summary_stats['LD_r_squared'] - LD_r_squared_mean
```

```python
summary_stats
```

```python
expected_beta = summary_stats['diff_beta'].sum()
expected_LD = summary_stats['diff_LD_r_squared'].sum()
```

```python
heritability = expected_beta *M / expected_LD *N
```

```python
heritability
```

```python
summary_stats['betaLD'] = summary_stats['beta_marginal'] * summary_stats['LD_r_squared']
```

```python
sum_beta_sq_ld = summary_stats['betaLD'].sum()
```

```python
sum_ld = summary_stats['LD_r_squared'].sum()
```

```python
summary_stats['LD_squared'] = summary_stats['LD_r_squared'] **2 
```

```python
sum_ld_sq = summary_stats['LD_squared'].sum()
```

```python
heritability = M * ((N * sum_beta_sq_ld - sum_ld) / (N * sum_ld_sq))
```

```python
heritability
```
